```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import ldl
from numpy.linalg import cholesky
```

```python
x = np.random.randn(2,5000)
```

```python
x.shape
```

```python
mean = np.array([0,0])
mean
```

```python
sigma = np.array([(10,0.5),(0.5,10)])
sigma
```

```python
#lu, d, perm = ldl(sigma)
```

```python
A = cholesky(sigma)
```

```python
A
```

```python
output = np.matmul(A.T,x)
```

```python
output.shape
```

```python
plt.figure(figsize=(8 ,8))
plt.scatter(output[0,:],output[1,:],ls='dotted',lw=1,color='blue',alpha=0.5)
plt.grid()
#plt.xlim([-4,4])
#plt.ylim([-4,4])
plt.title('Scatter plot',fontsize=16)
plt.xlabel('x1',fontsize=14)
plt.ylabel('x2',fontsize=14)
```
